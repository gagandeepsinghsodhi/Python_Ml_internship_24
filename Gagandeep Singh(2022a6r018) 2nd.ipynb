```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import os
%matplotlib inline
```

```python
task = pd.read_csv("titanic.csv")
task
```

```python
people= task.groupby(["sex","alive"]).size().unstack(fill_value = 0)
people

```

```python
people.plot(kind = "bar")
plt.grid()
plt.show()
```

```python
#Graph between people and deck
people= task.groupby(["sex","deck"]).size().unstack(fill_value = 0)
people
```

```python
people.plot(kind = "bar")
plt.grid()
plt.show()
```

```python
# Graph between class and pople
price= task.groupby(["class","who"]).size().unstack(fill_value = 0)
price
```

```python
price.plot(kind = "bar")
plt.grid()
plt.show()
```

```python
# Graph between Alone and alive and who
life= task.groupby(["alone","alive","who"]).size().unstack(fill_value = 0)
life

```

```python
life.plot(kind = "bar",color = ["Red", "Blue","Green"])
plt.grid()
plt.show()
```
